```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings as ignore
```

```python
airline_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Airline Dataset Updated - v2.csv')
df= airline_data
```

```python
df
```

```python
df.info()
```

```python
df.drop(["Passenger ID", "First Name", "Last Name", "Pilot Name"], axis = 1, inplace = True)
df
```

```python
df['Gender']=df['Gender'].astype('category')
df['Gender'].unique()
```

```python
df['Age']=df['Age'].astype(int)
```

# EDA


```python
_,ax=plt.subplots(figsize=(12,12), ncols=2, nrows=2)
columns=[key for key, value in df.nunique().to_dict().items() if value<10]
for index, column in enumerate(columns):
    df[column].value_counts().plot(ax=ax.ravel()[index], kind='bar', legend=True)
```

```python
Gender_count=df['Gender'].value_counts()

fig=px.pie(Gender_count, values=Gender_count.values, names=Gender_count.index, title="Distribution od passengers accross Gender")
fig.show()
```

```python
Country_Name_count=df['Country Name'].value_counts()
fig=px.bar(Country_Name_count.head(10), x=Country_Name_count.head(10).index, y=Country_Name_count.head(10).values,
           title='Top 10 Most Passenger Country',
           labels={'x':'Country', 'y':'Passenger Count'})
fig.show()
```

```python
fig = px.histogram(df, x='Age', nbins=20, color='Gender',
                   title='Histogram of Passengers Age',
                   labels={'Age': 'Passengers Age'})
fig.show()
plt.figure(figsize=(18,10))
sns.countplot(data=df,x='Age')
plt.xticks(rotation=90)
plt.show()
```

```python
fig = px.box(df, x='Gender', y='Age', color='Gender',
             title='Box Plot of Age by Gender',
             labels={'Age': 'Passenger Age'})
fig.show()
```

```python
fl_stat = pd.DataFrame(df['Flight Status'].value_counts()).reset_index()
```

```python
fig = px.pie(fl_stat, values='Flight Status', names='Flight Status', color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
print(fl_stat)
```

```python
status_counts = df.groupby(['Continents', 'Flight Status']).size().unstack()

status_proportion = status_counts.div(status_counts.sum(axis=1), axis=0)

status_proportion.plot(kind='bar', stacked=True, legend=True, color=('blue', 'orange', 'green'))
plt.show()
```
