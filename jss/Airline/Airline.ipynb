```python
def generation(age):
    if 1 <= age <= 10:
        return '10대 미만'
    elif 11 <= age <=20:
        return '10'
    elif 20 <= age < 30:
        return '20'
    elif 30 <= age < 40:
        return '30'
    elif 40 <= age < 50:
        return '40'
    elif 50 <= age < 60:
        return '50'
    elif 60 <= age < 70:
        return '60'
    else:
        return '70대 이상'

df['Generation'] = df['Age'].apply(generation)
```

```python
years, months, days = [], [], []

# Loop through each date in the 'Departure Date' column
for date in df['Departure Date']:
    if "-" in date:
        # Splitting the date using "-"
        parts = date.split("-")
    elif "/" in date:
        # Splitting the date using "/"
        parts = date.split("/")
    else:
        # Handling other date formats or providing a default value
        print(f"Unsupported date format: {date}")
        continue

    # Extracting Year, Month, and Day
    year, month, day = parts[2], parts[0], parts[1]

    # Append the values to the corresponding lists
    years.append(int(year))
    months.append(int(month))
    days.append(int(day))

# Add new columns to the DataFrame
df['Year'] = years
df['Month'] = months
df['Day'] = days
```

```python
# 각 Gender에서 국적 빈도수 계산
gender_nationality_counts = df.groupby(['Gender', 'Nationality'])['Nationality'].value_counts().reset_index(name='count')

# 남성과 여성 각각에 대해 상위 30개 국적 선택 및 정렬
male_top_nationalities = (
    gender_nationality_counts[gender_nationality_counts['Gender'] == 'Male']
    .nlargest(30, 'count')
    .sort_values(by='count', ascending=False)
)
female_top_nationalities = (
    gender_nationality_counts[gender_nationality_counts['Gender'] == 'Female']
    .nlargest(30, 'count')
    .sort_values(by='count', ascending=False)
)

# 그래프 크기 설정
plt.figure(figsize=(15, 8))

# 남성의 상위 30개 국적별 빈도수 시각화
plt.subplot(1, 2, 1)
sns.barplot(x='Nationality', y='count', data=male_top_nationalities)
plt.title('Top 30 Male Nationalities')
plt.xticks(rotation=90)

# 여성의 상위 30개 국적별 빈도수 시각화
plt.subplot(1, 2, 2)
sns.barplot(x='Nationality', y='count', data=female_top_nationalities)
plt.title('Top 30 Female Nationalities')
plt.xticks(rotation=90)

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
country_month_counts = df.groupby(['Country Name', 'Month'])['Country Name'].value_counts().reset_index(name='count')
```

```python
# 결과를 원하는 순서로 정렬
country_month_counts = country_month_counts.sort_values(by=['Month', 'count'], ascending=[True, False])

# 각 월별 상위 30개 나라 선택
top_countries_per_month = country_month_counts.groupby('Month').head(30)

# 월별 subplot 생성
months = top_countries_per_month['Month'].unique()
num_months = len(months)

plt.figure(figsize=(30, 25))

for i, month in enumerate(months, start=1):
    plt.subplot(3, 4, i)  # 3행 4열로 나누어진 subplot에서 i번째 위치
    data_month = top_countries_per_month[top_countries_per_month['Month'] == month]
    sns.barplot(x='Country Name', y='count', data=data_month, palette='viridis')
    plt.title(f'Top 30 Countries - {month}')
    plt.xlabel('Country Name')
    plt.ylabel('Count')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
