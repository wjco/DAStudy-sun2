# Hotel Bookings Analysis
 
- https://www.kaggle.com/datasets/thedevastator/hotel-bookings-analysis/data

#### 데이터셋 설명
- 호텔 예약, 취소, 투숙객에 관한 포괄적인 정보 모음입니다.
- 호텔 유형, 예약당 성인, 어린이 및 유아 수, 주말 밤(토요일 또는 일요일)과 주중(월~금)의 숙박 기간, 선택한 식사 계획 등 다양한 측면에 대한 통찰력을 제공합니다.

#### 데이터셋 구조
- hotel	호텔 유형(리조트 또는 도시)을 나타냅니다. (범주형)
- is_canceled	예약이 취소되었는지 여부를 지정합니다(0=취소되지 않음, 1=취소됨). (바이너리)
- lead_time	예약 날짜와 도착 날짜 사이의 일수를 나타냅니다. (숫자)
- arrival_date_year	도착일의 연도를 나타냅니다. (범주형)
- arrival_date_month	도착일의 월을 나타냅니다. (범주형)
- arrival_date_week_number	손님이 호텔에 도착한 주를 지정합니다. (숫자)
- arrival_date_day_of_month	한 달 내 특정 도착일을 나타냅니다. (숫자)
- stays_in_weekend_nights	주말 동안 호텔에 숙박하거나 예약한 투숙객의 숙박일수(토요일 또는 일요일)를 나타냅니다. (숫자)
- stays_in_week_nights	주중에 호텔에 숙박하거나 예약한 주중 밤(월요일~금요일) 수를 나타냅니다. (숫자)
- adults	각 예약에 포함된 성인의 수를 나타냅니다. (숫자)
- children	각 예약에 포함된 어린이의 수를 나타냅니다. (숫자)
- babies	각 예약에 포함된 아기의 수를 나타냅니다. (숫자)
- meal	예약된 식사 유형을 설명합니다(조식만 제공, 1박 2식, 3식 제공 또는 정의되지 않음/SC – 식사 패키지 없음). (범주형)
- country	예약한 각 고객의 출신 국가를 나타냅니다. (범주형)
- market_segment	예약 시 개인이 속하는 다양한 시장 부문을 표시합니다(예: 온라인 여행사, 오프라인 여행사, 기업 고객). (범주형)
- distribution_channel   유통 채널	예약이 이루어진 다양한 채널을 지정합니다(예: 온라인 여행사, 호텔/여행사를 통한 직접 예약, 기업 예약). (범주형)
- is_repeated_guest	손님이 반복 방문자인지 여부를 나타냅니다(0=반복 손님이 아님, 1=반복 손님). (바이너리)
- previous_cancellations	고객이 이전에 예약을 취소한 횟수를 나타냅니다. (숫자)
- previous_bookings_not_canceled	취소되지 않은 고객의 이전 예약 수를 나타냅니다. (숫자)
- reserved_room_type	처음에 예약된 객실 유형을 식별합니다. (범주형)
- assigned_room_type	손님에게 할당된 객실 유형을 식별합니다. (범주형)
- booking_changes	예약에 대한 변경 횟수를 나타냅니다. (숫자)
- deposit_type	예약에 대한 보증금 유형을 나타냅니다. (범주형)
- agent	예약한 여행사의 ID를 나타냅니다. (범주형)
- company	예약한 회사의 ID를 나타냅니다. (범주형)
- days_in_waiting_list	예약이 확인되기 전에 대기 목록에 있었던 일수를 나타냅니다. (숫자)
- customer_type	고객 유형(예: 임시, 계약, 그룹 또는 기타)을 나타냅니다. (범주형)
- ADR	예약에 대한 평균 일일 요금(객실당 가격)을 나타냅니다. (숫자)
- required_car_parking_spaces	손님이 요구하는 주차 공간의 수를 나타냅니다. (숫자)
- total_of_special_requests	고객이 요청한 총 특별 요청(예: 엑스트라 베드, 객실 편의 시설) 수를 나타냅니다. (숫자)
- reservation_status	예약 상태(예: 취소, 체크인, 노쇼)를 나타냅니다. (범주형)
- reservation_status_date	예약 상태가 마지막으로 업데이트된 날짜를 나타냅니다. (날짜)

## Data Visualization

#### 예약 취소 비율 확인

- 0 : 취소되지 않음 1 : 취소됨
- 취소된 비율보다 취소되지 않은 비율이 많음

```python
canceled_df = df['is_canceled'].value_counts(normalize =True)
canceled_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=canceled_df.index, y=canceled_df.values, palette='Set2')
plt.title('Cancellation Distribution')
plt.xlabel('Cancellation Status')
plt.ylabel('Percentage')
plt.show()
```

#### 호텔 유형(리조트 또는 도시) 확인

- 도시 호텔이 리조트보다 많음

```python
hotel_df = df['hotel'].value_counts()
hotel_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=hotel_df.index, y=hotel_df.values, palette='Set2')
plt.title('Hotels type')
plt.xlabel('Hotel reservation Status')
plt.show()
```

#### 호텔 유형별 취소 확인

- 도시 호텔에서 예약 비율만큼 취소 비율도 많은 것을 알 수 있음

```python
canceled_by_hotel = df.groupby('hotel')['is_canceled'].value_counts().reset_index(name='cancellation_count')

plt.figure(figsize=(10, 6))
sns.barplot(data=canceled_by_hotel, x='hotel', y='cancellation_count', hue='is_canceled', palette='Set2')
plt.title('Canceled Count by Hotel')
plt.xlabel('Hotel')
plt.show()
```

#### 각 월마다 호텔 예약 상태 확인

- 8월달에 예약이 가장 많은 것을 알 수 있음
- 12월 달이 예약이 가장 적은 것을 알 수 있음

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date']) #datetime으로 변환
df['month'] = df['reservation_status_date'].dt.month # month 추출

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='month', hue='is_canceled', palette='Set3')
plt.title('Reservation hotel per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(['not cancelled','cancelled'])
plt.xticks(rotation=90)
plt.show()
```

####  예약 평균 일일 요금(객실당 가격)

- 예약한 평균 일일 요금은 8월달이 가장 높음
- 취소된 예약건 중 1월달 요금이 가장 높았음

##### 예약건

```python
# 취소되지 않은 = 예약한 가격 월 평균
adr_per_month = df[df['is_canceled']==0].groupby('month')['adr'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=adr_per_month, x='month', y='adr')
plt.title('Average Daily Rate per Month')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

##### 취소건

- 예약이 취소될 때는 -> 숙박 비용이 가장 높을 때 많이 발생하며 비용이 가장 낮을 때 드물게 발생하는 것을 알 수 있음
- 이를 통해, 숙박 비용이 예약 취소에 유일하게 영향을 미치는 요인이라고 할 수 있음

```python
# 취소된 가격 월 평균
adr_per_month = df[df['is_canceled']==1].groupby('month')['adr'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=adr_per_month, x='month', y='adr')
plt.title('Average Daily Rate per Month')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

#### 각 변수들간 상관 관계 확인

**상관 관계가 높은 변수**
- arrival_date_week_number & arrival_date_year
    - 도착한 주와 도착한 연도
- statys_in_week_number & statys_in_week_nights
    - 손님이 호텔에 도착한 주와 주중에 호텔에 숙박하거나 예약한 투숙객의 숙박일수
- previous_bookings_not_canceled & is_repeated_guest
    - 취소되지 않은 고객의 이전 예약 수와 손님의 반복 방문자 여부 

```python
correlation_matrix = df.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", mask=mask)
plt.show()
```
