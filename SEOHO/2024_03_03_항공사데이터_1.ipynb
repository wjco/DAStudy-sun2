### 칼럼 설명
- Passenger ID - 각 승객의 고유 식별자
- First Name - 승객의 이름
- Last Name - 승객의 성
- Gender - 성별
- Age - 나이
- Nationality - 국적
- Airport Name - 공항명
- Airport Country Code - 공항 국가 코드
- Country Name - 공항의 국가명
- Airport Continent - 공항이 위치한 대륙
- Continents - 비행 경로에 포함된 대륙
- Departure Date - 출발 날짜
- Arrival Airport - 도착 공항명
- Pilot Name - 조종사 이름
- Flight Status - 항공편의 상태 (정시, 지연, 취소 등)

```python
df_Nationality =df[['Nationality','Flight Status']].value_counts().reset_index()
df_Nationality['total'] = ''
for i in df_Nationality['Nationality'].unique():
    df_Nationality.loc[df_Nationality['Nationality']==i,'total'] = df_Nationality.loc[df_Nationality['Nationality']==i][0].sum()
df_Nationality['rate'] = df_Nationality[0]/df_Nationality['total']
```

```python
df_Nationality[df_Nationality[0]>=30].sort_values('rate',ascending=False)
```

```python
df_Nationality[df_Nationality[0]>=30].sort_values(['Flight Status','rate'],ascending=[True,False]).head(20)
```

```python
df_Nationality[df_Nationality[0]>=30].sort_values(['Flight Status','rate'],ascending=[False,False]).head(20)
```
